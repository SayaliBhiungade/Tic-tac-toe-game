```python
from tkinter import *
from functools import partial
import numpy as np
a=Tk() 
a.title("Tik Tac Toe game")
a.geometry("300x300")
count=0
flag=0
click=0
win_combo=np.array([123,147,159,456,258,357,369,789])# winning combinations cells no. from 1 to 9
class X(Frame):
    
    def __init__(self,ro,col):
        Frame.__init__(self)
        self.grid()
        self.button_clicks=0
        self.create_widgets(ro,col)
        self.position(ro,col)
        #self.undo(click)
    def create_widgets(self,ro,col):
        self.button=Button(self)
        self.button.grid()
        self.set_text()
        self.button["height"]=3
        self.button["width"]=6
        self.button["command"]=partial(self.update_count,ro,col)
    def set_text(self):
        self.button["text"]=""
    def single_number(self,a,b,c):# to convert cells from 00 01 ,etc format to 1,2,etc. format
        a=3*a[0]+a[1]+1
        b=3*b[0]+b[1]+1
        c=3*c[0]+c[1]+1
        abc=np.array([a,b,c])
        abc=np.sort(abc)
        abc=100*abc[0]+10*abc[1]+abc[2]
        return abc
    def undo(self,r,c):
        global Undo,b,count,player1,player2
        Undo.config(state="disabled")
        b[r][c].set_text()
        self.button_clicks=0
        if(count%2==1):
            del player1[-1]
        else:
            del player2[-1]
        count-=1;
    def update_count(self,ro,col):
        global count,player1,player2,flag,win_combo,a,Undo,click
        self.button_clicks+=1
        Undo.config(command=partial(self.undo,ro,col))
        Undo.config(state=NORMAL)
        if(self.button_clicks==1):
            count+=1
            if(count%2!=0):
                sym=selection
                player1.append([ro,col])
            else:
                sym=sel2
                player2.append([ro,col])
            self.button["text"]=sym
        if((count>=5)and(flag==0)):
            t=np.array([[1,2,4],[1,3,4],[2,3,4]])
            if(count%2!=0):
                if(count==5):
                    check=self.single_number(player1[1],player1[2],player1[3])
                    comb1=np.array([check])
                    res1=np.in1d(comb1.ravel(), win_combo.ravel()).all()#check whether current combi is subset of win_combo or not
                    cell1=player1[1]
                    cell2=player1[2]
                    cell3=player1[3]
                else:
                    if(count==9):
                        t1=np.array([[1,2,5],[1,3,5],[1,4,5],[2,3,5],[2,4,5],[3,4,5]])
                        t=np.append(t,t1,axis=0)
                    for [i,j,k] in t:
                        check=self.single_number(player1[i],player1[j],player1[k])
                        comb1=np.array([check])
                        res1=np.in1d(comb1.ravel(), win_combo.ravel()).all()
                        if(res1):
                            cell1=player1[i]
                            cell2=player1[j]
                            cell3=player1[k]
                            break
                if(res1):
                    flag=1
                    #highlighting winning pattern
                    b[cell1[0]][cell1[1]].button["bg"]='yellow'
                    b[cell2[0]][cell2[1]].button["bg"]='yellow'
                    b[cell3[0]][cell3[1]].button["bg"]='yellow'
                    print("Winner:Player1")
                    textview.config(bg="#CCFFFF",fg="red")
                    textview.insert(END, "Player1")
                    #a.after(2500,lambda: a.destroy())
            else:
                if(count==6):
                    check=self.single_number(player2[1],player2[2],player2[3])
                    comb2=np.array([check])
                    res2=np.in1d(comb2.ravel(), win_combo.ravel()).all()
                    cell1=player2[1]
                    cell2=player2[2]
                    cell3=player2[3]
                else:
                    for [i,j,k] in t:
                        check=self.single_number(player2[i],player2[j],player2[k])
                        comb2=np.array([check])
                        res2=np.in1d(comb2.ravel(), win_combo.ravel()).all()
                        if(res2):
                            cell1=player2[i]
                            cell2=player2[j]
                            cell3=player2[k]
                            break
                if(res2):
                    flag=2
                    b[cell1[0]][cell1[1]].button["bg"]='yellow'
                    b[cell2[0]][cell2[1]].button["bg"]='yellow'
                    b[cell3[0]][cell3[1]].button["bg"]='yellow'
                    print("Winner:Player2")
                    textview.config(bg="#CCFFFF",fg="red")
                    textview.insert(END, "Player2")
                    #a.after(2500,lambda: a.destroy())
            if(count==9 and flag==0):
                print("Result of the game is draw")
                textview.delete('1.0', END)
                textview.config(bg="#CCFFFF",fg="red")
                textview.insert(END, "Result of the game is draw")
                #a.after(2500,lambda: a.destroy())
    def position(self,ro,col):
        self.grid(row=ro,column=col)
rows, cols = (3, 3) 
b = [[0 for i in range(cols)] for j in range(rows)] 
player1=[[]]
player2=[[]]
#Creating GUI of 9 buttons in 3x3 pattern
for i in range(0,3):
    for j in range(0,3):
        b[i][j]=X(i,j)

Undo=Button(a,text="Undo",fg="blue")
#Undo=X(1,6)
Undo.place(x=200,y=70)
textview=Text(a,height=2,width=28, relief=GROOVE,font='Georgia 10 bold')#textview
textview.place(x=2,y=180)
textview.insert(INSERT, "Winner is .....")
selection=input("Choose the symbol --->Player 1: enter X or O ==>> ")
if(str(selection)=="X"):
    sel2="O"
else:
    sel2="X"
print("Player 2 symbol:",sel2)
a.mainloop()
```
